```python
import pandas as pd

df = pd.read_csv(r'C:\\Users\\Admin\\Downloads\\OnlineRetail.csv',encoding= 'unicode_escape')
df
```

```python
df.info()
```

```python
df['Quarter']= pd.to_datetime(df['InvoiceDate']).dt.month
df
```

```python
df['Quarter'].replace([1,2,3],1,inplace = True)
df['Quarter'].replace([4,5,6],2,inplace = True)
df['Quarter'].replace([7,8,9],3,inplace = True)
df['Quarter'].replace([10,11,12],4,inplace = True)
df
```

```python
df['Amount'] = df['Quantity'] * df['UnitPrice']
df
```

```python
import numpy as np
```

```python
ten = (df['Country'] == 'United Kingdom') & (df['Quarter'] == 4)
five = df['Country'] == 'France'

conditions = [ten,five]
value = [10 % ; 5 %]

df['Discount'] = np.select(conditions, value, default = 0 %)
df
```

```python
df['Discount'] = pd.to_numeric(df['Discount'])
```

```python
df['Total'] = df['Amount'] * (1 - df['Discount'])
df
```
